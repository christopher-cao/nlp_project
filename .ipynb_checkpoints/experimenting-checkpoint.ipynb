```python
# Function that returns a list that says exactly which words/phrases were found that triggered the suspicious flag
def findNegativeOverlap(full_text):
    temp = []
    for i in negative_words:
        if i in full_text:
            temp.append(i)
    return temp

def findPositiveOverlap(full_text):
    temp = []
    for i in positive_words:
        if i in full_text:
            temp.append(i)
    return temp
```

```python
def articleFeatures(index, df):
    features = {}
    features['positive_words'] = tuple(df.ix[index]['POSITIVE'])
    features['negative_words'] = tuple(df.ix[index]['REASON'])
    return features

def articleClass(index, df):
    return df.ix[index]['ANNOUNCEMENT']

articleFeatures(3, final_train)
```

nb = Naive Bayes

```python
# NEED TO CONVERT IT TO TUPLES 
# (article, classification) e.g. (6, Y) 
example = (6, articleClass(6, final_train))

# Got tuples with the correct classification ! !!!!!
nb_train = [(articleFeatures(index, final_train), articleClass(index, final_train)) for index,rows in final_train.iterrows()]
# nb_test = [(index,)]
nb_test = [(articleFeatures(index, final_test), articleClass(index, final_test)) for index, rows in final_test.iterrows()]
```

```python
classifier = nltk.NaiveBayesClassifier.train(nb_train)
```

So turns out this classifier actually performs better than the other one in that it actually gives some positive guesses

```python

guesses = []

for index, row in final_test.iterrows():
    guess = "Guess: " + str(classifier.classify(articleFeatures(index, final_test)))
    correct_tag = "Correct: " + str(articleClass(index, final_test))
    negative_words = "-: " + str(final_test.ix[index]['REASON'])
    positive_words = "+: " + str(final_test.ix[index]['POSITIVE'])
    guesses.append( (index, correct_tag, guess, negative_words, positive_words))
```

Maybe what I should do is make the features just the existence of a word in the fulltext. Go through every word in the negative words list and assign a 0 or 1 based on if its in the full text. Do the same for positive words.
- 'World Bank': 0/1

```python
# Need to find a more efficient way to do this... maybe tokenizing it will help
def articleFeatures2(index, df):
    features = {}
    for word in negative_words:
        features['contains ({}) (-)'.format(word)] = word in df.ix[index]['FULLTEXT']
    for word in positive_words:
        features['contains ({}) (+)'.format(word)] = word in df.ix[index]['FULLTEXT']
    return features
```

```python
# Got tuples with the correct classification ! !!!!!
nb_train2 = [(articleFeatures2(index, final_train), articleClass(index, final_train)) for index,rows in final_train.iterrows()]
# nb_test = [(index,)]
nb_test2 = [(articleFeatures2(index, final_test), articleClass(index, final_test)) for index, rows in final_test.iterrows()]
```

```python
classifier2 = nltk.NaiveBayesClassifier.train(nb_train2)
```

```python
print(nltk.classify.accuracy(classifier2, nb_test2))
```

```python
classifier2.show_most_informative_features(40)
```

Now for the <b>error analysis</b>

```python
errors = []

for index, row in final_test.iterrows():
    guess = classifier2.classify(articleFeatures2(index, final_test))
    correct_tag = articleClass(index, final_test)
    if guess != correct_tag:
        negative_words = "-: " + str(final_test.ix[index]['REASON'])
        positive_words = "+: " + str(final_test.ix[index]['POSITIVE'])
        errors.append( (index, correct_tag, guess, negative_words, positive_words))
```

```python
def keepRelevant(df):
    temp = []
    for index, rows in df.iterrows():
        announcement = df.iloc[index]['ANNOUNCEMENT']
        if announcement == 'N' or announcement == 'Y':
            temp.append(index)
    return df.ix[temp]
```

```python
def getWords(file_name):
    fin = open(file_name, 'r')
    return [i.strip() for i in fin]
```

```python
def findNegativeOverlap(full_text):
    temp = []
    for i in negative_words:
        if i in full_text:
            temp.append(i)
    return temp

def findPositiveOverlap(full_text):
    temp = []
    for i in positive_words:
        if i in full_text:
            temp.append(i)
    return temp
```

```python
negative_words = getWords('project_initial_data/negative_words.txt')
positive_words = getWords('project_initial_data/positive_words.txt')
```
